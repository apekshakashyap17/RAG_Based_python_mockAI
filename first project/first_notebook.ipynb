```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


textsplitter = RecursiveCharacterTextSplitter(
    chunk_size = 800,
    chunk_overlap = 120,

     separators=[
        "\n\n",   # paragraphs
        "\n",     # lines
        ". ",     # sentences
        "? ",
        "! ",
        ", ",
        " ",
        ""
    ]
)   

chunks = textsplitter.split_documents(data)

print(len(chunks))
print(chunks[3].page_content[:])
```

```python
import random

for i, c in enumerate(random.sample(chunks, 7)):
    print(f"\n\n========== CHUNK {i} ==========")
    print(c.page_content[:800])


```
